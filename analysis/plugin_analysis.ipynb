```python
import pandas as pd
import os

from utils import load_scores, get_achieved_best_plugin
```

```python
meta_model = 'sl'
base_models = ['l1', 'l2', 'dt', 'rf', 'et', 'kr', 'cb', 'lgbm']
est_dir = '../results/metrics/run1/'


scorer_dir = '../results/scores/run1/'
scorer_est = 'sl_dt'
path = f'{scorer_dir}{scorer_est}'
scorer_type = 'plugin'

best_list = []
for bm in base_models:
    est_name = f'{meta_model}_{bm}'
    df_test = pd.read_csv(os.path.join(est_dir, est_name, f'{est_name}_test_metrics.csv'))
    df_plugin = pd.read_csv(os.path.join(scorer_dir, scorer_est, f'{est_name}_{scorer_type}_{scorer_est}.csv'))

    main_df = load_scores(df_plugin, df_test)

    best_i = get_achieved_best_plugin(main_df)
    best_list.append([est_name] + best_i)
```

```python
df_best = pd.DataFrame(best_list, columns=['name', 'ate_on_plugin_ate', 'ate_on_plugin_pehe', 'ate_best', 'pehe_on_plugin_ate', 'pehe_on_plugin_pehe', 'pehe_best'])
```
