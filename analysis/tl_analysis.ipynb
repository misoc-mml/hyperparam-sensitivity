```python
import pandas as pd

from utils import load_merge, get_mse_corr, get_risk
```

```python
path = '../results/metrics/'

meta_model = 'tl'
base_models = ['l1', 'l1_inter', 'l2', 'l2_inter', 'dt', 'rf', 'et', 'kr', 'cb', 'lgbm']

mse_corr_list = []
risk_list = []
```

```python
for b_m in base_models:
    main_df = load_merge(path, f'{meta_model}_{b_m}', cols=['mse_m0', 'mse_m1'])

    corr_i = get_mse_corr(main_df)
    mse_corr_list.append([f'{meta_model}-{b_m}'] + corr_i)

    r_ate, r_pehe = get_risk(main_df)
    risk_list.append([f'{meta_model}-{b_m}'] + r_ate + r_pehe)
```

```python
df_corr = pd.DataFrame(mse_corr_list, columns=['name', 'ate_val', 'pehe_val', 'mse_test', 'ate_test', 'pehe_test'])
```

```python
df_risk = pd.DataFrame(risk_list, columns=['name', 'ate_risk', 'pehe_risk'])
```
